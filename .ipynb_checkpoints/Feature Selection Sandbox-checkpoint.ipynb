```python
import operator
term_indices = list(vec.vocabulary_.items())
term_indices.sort(key=operator.itemgetter(1))

term_list = [i[0] for i in term_indices]
```

```python
data = adjusted.toarray()
```

```python
[i[91] for i in data][:10]
```

```python
from scipy.stats.stats import pearsonr
p_tuples = []
for column in data.T:
    p_c = pearsonr(column, dates)
    p_tuples.append(p_c)

print(len(term_list), len(p_tuples))
```

```python
final_tuples = list(zip(term_list, [i[0] for i in p_tuples], [format(i[1], '.14f') for i in p_tuples]))
final_tuples[1]
```

```python
df = pd.DataFrame(final_tuples, columns=["term", "pearson", "p_value"])
df.sort(columns="pearson", ascending=False)
```
